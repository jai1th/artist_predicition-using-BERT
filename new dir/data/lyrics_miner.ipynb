```python
import os
import json
```

```python
def clean_lyrics_by_rule(text):
    lines = text.split('\n')

    # Removing the second line after the title, which usually contains garbage string like username and year submitted or id
    if len(lines) > 1:
        lines.pop(1)

    cleaned_lines = []
    for line in lines:
        stripped = line.strip()

        # Skipping section headers like [Verse 1]
        if stripped.startswith('['):
            continue
        # Skipping empty lines
        if not stripped:
            continue
        stripped = stripped.replace('\u2014', '—')

        cleaned_lines.append(stripped)

    return "\n".join(cleaned_lines)
```

```python
def convert_lyrics_to_json(folder_path):
    data = []
    for file in os.listdir(folder_path):
        if file.endswith(".txt"):
            label = file.replace("_lyrics.txt", "").replace("_", " ").title()
            with open(os.path.join(folder_path, file), "r", encoding="utf-8") as f:
                songs = f.read().split("===\n\n")
                for song in songs:
                    text = clean_lyrics_by_rule(song)
                    # if len(text.split()) >= 20 and text[:3].isnumeric():  # To skip the short stuff
                    # if len(text.split()) >= 20:
                    # text = text.replace('\u2014', '—')
                    print(text)
                    data.append({"text": text, "label": label})
    return data

dataset = convert_lyrics_to_json("../scraping/lyrics")
with open("lyrics_dataset.json", "w", encoding="utf-8") as out:
    json.dump(dataset, out, indent=2, ensure_ascii=False)
```
